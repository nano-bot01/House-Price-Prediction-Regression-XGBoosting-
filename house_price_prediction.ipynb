Importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset


```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
# now we are making structured data using pandas library
house_price_dataframe = pd.DataFrame(house_price_dataset.data)
house_price_dataframe
```

```python
# inserting column names using feature_names on house_price_dataset
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
house_price_dataframe.head()
```

```python
# add the target column to the dataframe
# prices are target column here 
house_price_dataframe['price'] = house_price_dataset.target
house_price_dataframe
```

```python
# finding the number of rows and columns in dataset 
house_price_dataframe.shape
```

```python
# processing for missing/ wrong/ Null values
# there is no missing values

house_price_dataframe.isnull().sum()
```

```python
# statistical measures of dataset
house_price_dataframe.describe()
# describe function is like summary which gives mean, count, min, max, 
# 25th, 50th, 75th quartile and standard deviation.
```

## Understanding and ploting the Correlation between various features in the dataset 
> there are two types of correlation 
* Positive: is one increases other also increases.
* Negative correlation : If one increasese other decreases.

```python
# finding correlation between all data columns
correlation = house_price_dataframe.corr()

correlation
```

```python
# constructing heatmap to understand the correlation 
plt.figure(figsize= (10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot= True, annot_kws={'size' : 8}, cmap='Reds') 

# correlation is variable for corr()
# cbar is vertical bar for referencing
# square = all value must shown in square block
# fmt : minimum float value to represent a block
# annot : annotation are true, written on heatmap
# annot_kws : size of text on heatmap 
# cmap : color of map ( Greens, Reds, Blues, afmhot, 'Oranges_r', 'PRGn', 'PRGn_r', 'Paired', 'Paired_r', 'Pastel1', 'Pastel1_r', 'Pastel2', 'Pastel2_r', 'PiYG')
```

```python
# creating testing data and price variable
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
# printing both variables
print(X.head())
print(Y.head())
```

## Now, we are splitting data into two sets 
* Training data 
* Testing data 


```python
# train_test_split splits data for training and testing 
# test_size : ratio of traing and testing data split (here, 80% is training data )
# random_state : data splits in same way 

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state= 2)
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape )
```

# Model Training 
## XGBoost Regression
here we are using more than one machile learning models


```python
# loading model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

# Evaluation 
### we can't find directly predicted values but can find mean squared error or absolute errors

## Prediction on training data

```python
# accuracy fo rpredicton on traning data
# now we will compare predicted values to Y_train values

tr_prediction = model.predict(X_train)
```

```python
print(tr_prediction)
```

```python
# R squared error 
score_1 = metrics.r2_score(Y_train, tr_prediction)
print("R squared error ",score_1)


# MEAN Absolute error (MAE)
score_2 = metrics.mean_absolute_error(Y_train, tr_prediction)
print("Mean Absolute squared error ",score_2)
```

## Prediction on test data

## Visualizing the actual prices and predicted prices

```python
# visualizing actual prices vs predicted prices for training data
plt.scatter(Y_train, tr_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price VS Predicted Price")
plt.show()
```

```python
test_prediction = model.predict(X_test)
print(test_prediction)
```

```python
# R squared error for test data
score_3 = metrics.r2_score(Y_test, test_prediction)
print("R squared error ",score_3)


# MEAN Absolute error (MAE)
score_4 = metrics.mean_absolute_error(Y_test, test_prediction)
print("Mean Absolute squared error ",score_4)
```

```python
# visualizing actual prices vs predicted prices for testing data

plt.scatter(Y_test, test_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price VS Predicted Price")
plt.show()
```
